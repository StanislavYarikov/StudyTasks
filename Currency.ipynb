```python
import pandas as pd
import numpy as np
from datetime import datetime, date, time

#cur = pd.DataFrame(columns=['date', 'currency', 'operation', 'value', 'rate'])
cur = pd.read_csv('./currancy_PS.csv')

def add_cur (date, currency, operation, value, rate):
    global cur
    ADD_df = pd.DataFrame(data={'date': [date], 'currency': [currency], 'operation': [operation], 'value': [value], 'rate': [rate]})
#    cur2 = cur.append(ADD_df, ignore_index=True)
    cur = pd.concat([cur, ADD_df], ignore_index=True, sort=False)
    cur = cur.to_csv('./currancy_PS.csv', index=False)
    cur = pd.read_csv('./currancy_PS.csv')
    return cur

def EUR_buy (value, rate):
    global cur
    ADD_df = pd.DataFrame(data={'date': [tday], 'currency': ['EUR'], 'operation': ['buy'], 'value': [value], 'rate': [rate]})
    cur = pd.concat([cur, ADD_df], ignore_index=True, sort=False)
    cur = cur.to_csv('./currancy_PS.csv', index=False)
    cur = pd.read_csv('./currancy_PS.csv')
    return cur
def EUR_sell (value, rate):
    global cur
    ADD_df = pd.DataFrame(data={'date': [tday], 'currency': ['EUR'], 'operation': ['sell'], 'value': [value], 'rate': [rate]})
    cur = pd.concat([cur, ADD_df], ignore_index=True, sort=False)
    cur = cur.to_csv('./currancy_PS.csv', index=False)
    cur = pd.read_csv('./currancy_PS.csv')
    return cur

def USD_buy (value, rate):
    global cur
    ADD_df = pd.DataFrame(data={'date': [tday], 'currency': ['USD'], 'operation': ['buy'], 'value': [value], 'rate': [rate]})
    cur = pd.concat([cur, ADD_df], ignore_index=True, sort=False)
    cur = cur.to_csv('./currancy_PS.csv', index=False)
    cur = pd.read_csv('./currancy_PS.csv')
    return cur
def USD_sell (value, rate):
    global cur
    ADD_df = pd.DataFrame(data={'date': [tday], 'currency': ['USD'], 'operation': ['sell'], 'value': [value], 'rate': [rate]})
    cur = pd.concat([cur, ADD_df], ignore_index=True, sort=False)
    cur = cur.to_csv('./currancy_PS.csv', index=False)
    cur = pd.read_csv('./currancy_PS.csv')
    return cur

def del_str(number):
    global cur
    cur = cur.drop(cur.index[number])
    cur = cur.reset_index()
    del cur['index']
    cur = cur.to_csv('./currancy_PS.csv', index=False)
    cur = pd.read_csv('./currancy_PS.csv')
    return cur




tday = datetime.now().date()
cur
```

```python
cur_sum = cur.groupby(['currency', 'operation', 'rate'], as_index=False).agg({'value': 'sum'})
cur_sum['RUR'] = cur_sum.rate*cur_sum.value
cur_sum
```

```python
cur_svod = cur_sum.groupby(['currency', 'operation'], as_index=False).agg({'value': 'sum', 'RUR': 'sum'})
cur_svod['mean_rate'] = cur_svod.RUR/cur_svod.value
cur_svod
```

```python
cur_svod_have = cur_svod.drop('mean_rate', axis=1)
cur_svod_have.loc[cur_svod_have['operation']=='sell', 'RUR'] *= (-1)
cur_svod_have.loc[(cur_svod_have['operation']=='sell'), 'value'] *= (-1)
cur_svod_have
```

```python
cur_svod_have = cur_svod_have.groupby('currency', as_index=False).agg({'value': 'sum', 'RUR': 'sum'})
print('В наличии:')
cur_svod_have
```

```python
td_rate_USD = 74.10
td_rate_EUR = 89.15
n_max = len(cur_svod_have.index)

for i in range(n_max):
    if cur_svod_have.currency[i] == 'EUR':
        print('сейчас имеется', cur_svod_have.value[i], 
              'евро, можно продать за', cur_svod_have.value[i]*td_rate_EUR,
             'и получить прибыль', round(cur_svod_have.value[i]*td_rate_EUR-cur_svod_have.RUR[i], 2), 'руб.')
    elif cur_svod_have.currency[i] == 'USD':
        print('сейчас имеется', cur_svod_have.value[i], 
              'долларов, можно продать за', cur_svod_have.value[i]*td_rate_USD,
             'и получить прибыль', round(cur_svod_have.value[i]*td_rate_USD-cur_svod_have.RUR[i], 2), 'руб.')    
    else: print('имеется валюта', cur_svod_have.currency[i], 'в количестве ',  cur_svod_have.value[i])
    
    
```
